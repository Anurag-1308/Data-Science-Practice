**Indexing And Slicing**

```python
import numpy as np
a1= np.arange(10)
a2= np.arange(12).reshape(3,4)
a3= np.arange(8).reshape(2,2,2)
a1
```

```python
print(a1[-1])
print(a1[0])

print(a2[1,2])
print(a2[2,3])
print(a2[1,0])

print(a3[1,0,1])
print(a3[0,1,0])
print(a3[0,0,0])
```

```python
#1d matrix slicing

print(a1[2:5])

#2d matrix slicing

print(a2[0,:])
print(a2[:,2])
print(a2[1:,1:3])
print(a2[::2,::3])
print(a2[0:2,1:])

#3d matrix slicing
a3= np.arange(27).reshape(3,3,3)

print(a3[1])
print(a3[0,1,:])
print(a3[1,:,1])
print(a3[2,1:,1:])
print(a3[::2,0,::2])

```

**Iterating**

```python
for i in a1:
  print(i)

for i in a2:
  print(i)

for i in a3:
  print(i)

for i in np.nditer(a3):
  print(i)
```

**Stacking**

```python
#hprizontal stacking
a4= np.arange(12).reshape(3,4)
a5= np.arange(12,24).reshape(3,4)

print(np.hstack((a4,a5)))
print(np.hstack((a4,a5,a4,a5)))
```

```python
#vertical stacking
print(np.vstack((a4,a5)))
print(np.vstack((a4,a5,a4,a5)))
```
