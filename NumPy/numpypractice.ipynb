```python
#np.array
import numpy as np

a=np.array([1,2,3])
print(a)
b= np.array([[1,2,3], [4,5,6]])
print(b)
```

```python
#3d array
c= np.array ([[[1,2,3],[4,5,6]],[[6,7,8],[9,0,1]]])
print(c)
```

```python
# d type
np.array([1,2,3], dtype= bool)
np.array([1,2,3], dtype= float)
np.array([1,2,3], dtype= complex)
```

```python
#np.arange
np.arange(1,11)
np.arange(1,11,2)
```

```python
#np.reshape
np.arange(1,11).reshape(2,5)
```

```python
#ones and zeros
np.ones((3,4))
np.zeros((3,4))
```

```python
#np.random
np.random.random((1,2))
```

```python
#np.linspace
np.linspace(-10,10,10)
```

```python
#np.identity
np.identity(3)
```

```python
a1= np.arange(10)
a2= np.arange(12,dtype= float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
#ndim
a1.ndim
a2.ndim
a3.ndim
```

```python
#shape
a1.shape
a2.shape
a3.shape
```

```python
#size
a1.size
a2.size
a3.size
```

```python
#itemsize
a1.itemsize
a2.itemsize
a3.itemsize
```

```python
#dtype
a1.dtype
a2.dtype
a3.dtype
```
