# **Importing Pandas**

```python
import numpy as np
import pandas as pd

marks= [50,90,20,85]
subjects= ['maths', 'english','hindi', 'science']
```

# **Creating series**

```python
#using list
marks_series= pd.Series(marks)
print(marks_series)

#using custom index
print(pd.Series(marks,index=subjects))

#setting a name
print(pd.Series(marks,index=subjects,name= "Anurag's Marks"))

#using Dictionary
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}
print(pd.Series(marks))
```

# **SERIES ATTRIBUTES**

```python
#SIZE/dtype/name/index/value/is_unique
s= pd.Series(marks)

print(s.size)
print(s.dtype)
print(s.name)
print(s.index)
print(s.values)
print(s.is_unique)

```
